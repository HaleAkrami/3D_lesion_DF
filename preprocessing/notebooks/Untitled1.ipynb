```python
import pandas as pd
import shutil
import os



# Create the T1 directory if it doesn't exist
dir = '/scratch1/akrami/Latest_Data/camcan_preproc/'
file_path  = os.listdir(dir) 
out_path  = '/scratch1/akrami/Latest_Data/Data_pre/CamCan_pre/t2/'
# Process each file path
for file in file_path:
    # Check if the file exist
    #print(file)
    file_dir = dir+file+'/T2mni.nii.gz'
    print(file_dir)
    if os.path.exists(file_dir ):
        
        # Extract the subname from the file path
        # Extract only the desired part of the subname (before the first underscore)
        #main_subname = full_subname.split('_')[0]
        # Define the destination path
        dest_path_sub = f'{file}_t2.nii.gz'
        dest_path = out_path+ dest_path_sub
        #print(dest_path)
        # Copy the file to the destination
        shutil.copy(file_dir,dest_path)
```
