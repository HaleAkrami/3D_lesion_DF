```python
import filecmp
import os
import shutil

def compare_and_copy(source_dir, target_dir, new_dir):
    # Create the new directory if it doesn't exist
    if not os.path.exists(new_dir):
        os.makedirs(new_dir)

    # Get the list of files in the source directory
    source_files = os.listdir(source_dir)
    target_files = os.listdir(target_dir)

    # Iterate over the files in the source directory
    for file in source_files:
        # Modify the file name to match the target format
        source_file = file.replace("-T2", "_t2")
        #print(source_file)
        
        # Compare the modified file in the source directory with the target file
        
        if source_file not in target_files:
            source_files = file.replace("_t2","-T2")
            # Construct the source and target file paths
            new_file_path = os.path.join(new_dir, source_files)
            source_file_path = os.path.join(source_dir, source_files)
            #print(source_file_path)
            shutil.copy2(source_file_path, new_file_path)
            print(f"Copied {source_file_path} to {new_file_path}")

# Usage example
source_directory = "/scratch1/akrami/Data_pre/IXI/t2"
target_directory = "/scratch1/akrami/Data/Train/ixi/t2"
new_directory = "/scratch1/akrami/remain_IXI"

compare_and_copy(source_directory, target_directory, new_directory)
```
