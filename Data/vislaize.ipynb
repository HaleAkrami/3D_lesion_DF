```python
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np

# Load the .nii.gz file
nii_file_path = './SVRegTestData/Test_t1/t1/sub1_T1w_t1.nii.gz'
img = nib.load(nii_file_path)

# Convert to numpy array
data = img.get_fdata()

# You can specify which slice of the scan to visualize.
# For instance, to visualize the middle slice in the axial plane, you can do:
slice_index = data.shape[-1] // 2  # Assuming the last dimension is the axial plane

# Display the slice
plt.imshow(data[:, :, slice_index], cmap='gray')
plt.axis('off')  # To turn off the axis
plt.title('Axial Slice')
plt.show()
```
